```python
u=pd.read_csv('cleaned udemy.csv')
u
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='price', y='num_subscribers', data=u)
plt.title('Scatter Plot of Price vs Number of Subscribers')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(u['num_reviews'], bins=30, kde=True)
plt.title('Histogram of Number of Reviews')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='level', y='price', data=u)
plt.title('Box Plot of Price by Level')
plt.show()
```

```python
# Price vs Number of Subscribers
plt.figure(figsize=(10, 6))
sns.relplot(x='price', y='num_subscribers', data=u)
plt.title('Relational Plot of Price vs Number of Subscribers')
plt.show()
```

```python
sns.jointplot(x='price', y='num_reviews', data=u, kind='reg')
plt.suptitle('Joint Plot of Price vs Number of Reviews')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.regplot(x='num_lectures', y='num_reviews', data=u)
plt.title('Regression Plot of Number of Lectures vs Number of Reviews')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.regplot(x='price', y='num_reviews', data=u)
plt.title('Regression Plot of Price vs Number of Reviews')
plt.show()
```

```python
# Price by Level
plt.figure(figsize=(10, 6))
sns.violinplot(x='level', y='price', data=u)
plt.title('Violin Plot of Price by Level')
plt.show()
```

```python
# Course Level
plt.figure(figsize=(10, 6))
sns.countplot(x='level', data=u)
plt.title('Count Plot of Course Level')
plt.show()
```

```python
#Average Number of Subscribers by Level
plt.figure(figsize=(10, 6))
sns.barplot(x='level', y='num_subscribers', data=u)
plt.title('Bar Plot of Average Number of Subscribers by Level')
plt.show()
```

```python
sns.pairplot(u)
```

```python
# Price vs Number of Subscribers
plt.figure(figsize=(10, 6))
plt.scatter(u['price'], u['num_subscribers'])
plt.title('Scatter Plot of Price vs Number of Subscribers')
plt.xlabel('Price')
plt.ylabel('Number of Subscribers')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(u['level'], u['num_subscribers'])
plt.title('Bar Plot of Average Number of Subscribers by Level')
plt.xlabel('Level')
plt.ylabel('Number of Subscribers')
plt.xticks(rotation=50)
plt.show()
```

```python
#Number of Reviews
plt.figure(figsize=(10, 6))
plt.hist(u['num_reviews'], bins=30, edgecolor='black')
plt.title('Histogram of Number of Reviews')
plt.xlabel('Number of Reviews')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.boxplot(u['price'])
plt.title('Box Plot of Price')
plt.ylabel('Price')
plt.grid(True)
plt.show()
```

```python
avg= u.groupby('level')['num_subscribers'].mean()
plt.figure(figsize=(10, 6))
avg.plot(kind='bar')
plt.title('Average Number of Subscribers by Course Level')
plt.xlabel('Course Level')
plt.ylabel('Average Number of Subscribers')
plt.xticks(rotation=45)
plt.show()
```

```python
courses_by_level = u['level'].value_counts()
plt.figure(figsize=(10, 6))
courses_by_level.plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Courses by Level')
plt.axis('equal')  
plt.show()
```
