# decision tree

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
from sklearn.metrics import r2_score,accuracy_score,confusion_matrix,precision_score,classification_report
x=u.iloc[:,:-1]
x
```

```python
y=u['GradeClass']
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
print(x_train,x_test,y_train,y_test)
```

```python
tree_model=DecisionTreeClassifier(criterion='entropy',max_depth=3,random_state=42) 
tree_model.fit(x_train,y_train)
```

```python
pred=tree_model.predict(x_test)
pred
```

```python
dt_accuracy=accuracy_score(y_test,pred)
print(dt_accuracy)
```

```python
cnf=confusion_matrix(y_test,pred)
print(cnf)
```

```python
pre=precision_score(y_test,pred,average='macro')
print(pre)
```

```python
print(classification_report(y_test,pred))
```

```python
ne=np.array([[1001,17,1,0,2,19.833723,7,1,2,0,0,1,0,2.929196]])
ps=tree_model.predict(ne)
print(f'GradeClass: {ps[0]:.2f}')
```

```python
plt.figure(figsize=(40,20))
plot_tree(tree_model,
          filled=True,
          feature_names=x.columns,
          class_names=[str(cls) for cls in tree_model.classes_],
          rounded=True)
plt.title("decision Tree visualisation")
plt.show()
          
```

```python
from sklearn.cluster import KMeans
```

```python
k=3
```

```python
kmeans=KMeans(n_clusters=k,random_state=0) #n_clusters- no of clusters you want to genrate
kmeans.fit(x)
```

```python
label=kmeans.labels_ 
u['cluster']=label     
u['cluster']            
```

```python
u
```

```python
plt.figure(figsize=(10,6))
colors=plt.get_cmap('viridis',k)
for cluster in range(k):  
    cd=u[u['cluster']==cluster]
    plt.scatter(cd['Age'],cd['Gender'],color=colors(cluster),label=f'cluster {cluster + 1}')
centroids=kmeans.cluster_centers_ 
plt.scatter(centroids[:,0],centroids[:,1],color='black',marker='x',s=100,label='centroids')
plt.xlabel('Age')
plt.ylabel('Gender')
plt.title('clusters of age and gender')
plt.legend()
plt.show()
```
